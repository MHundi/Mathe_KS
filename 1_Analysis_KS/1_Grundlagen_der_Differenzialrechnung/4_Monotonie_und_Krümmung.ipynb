**Definition:**

Gilt für alle $x_1, x_2 \in \mathbb{D}$, wobei $x_1 < x_2$ ist, auch dass $f(x_1) < f(x_2)$ ist, dann heißt f **streng monoton wachsend (steigend) auf $\mathbb{D}$**

Gilt für alle $x_1, x_2 \in \mathbb{D}$, wobei $x_1 < x_2$ ist, auch dass $f(x_1) > f(x_2)$ ist, dann heißt f **streng monoton fallend auf $\mathbb{D}$**

Manchmal ist Monotnoie nur auf Intervallen der Funktion vorhanden. Man kann analog definieren:

Gilt für alle $x_1, x_2 \in I_1$, wobei $x_1 < x_2$ ist, auch dass $f(x_1) < f(x_2)$ ist, dann heißt f **streng monoton wachsend (steigend) auf $I_1$**

Gilt für alle $x_1, x_2 \in I_2$, wobei $x_1 < x_2$ ist, auch dass $f(x_1) > f(x_2)$ ist, dann heißt f **streng monoton fallend auf $I_2$**
</div>

Auf anaolger Wiese lässt sich auch die nichtstrenge Monotonie definieren.

<div style="border: 1px solid blue">


**Definition:**

Gilt für alle $x_1, x_2 \in \mathbb{D}$, wobei $x_1 < x_2$ ist, auch dass $f(x_1)\leq f(x_2)$ ist, dann heißt f **monoton wachsend (steigend) auf $\mathbb{D}$**

Gilt für alle $x_1, x_2 \in \mathbb{D}$, wobei $x_1 < x_2$ ist, auch dass $f(x_1)\geq f(x_2)$ ist, dann heißt f **monoton fallend auf $\mathbb{D}$**

Manchmal ist Monotnoie nur auf Intervallen der Funktion vorhanden. Man kann analog definieren:

Gilt für alle $x_1, x_2 \in I_1$, wobei $x_1 < x_2$ ist, auch dass $f(x_1) \leq f(x_2)$ ist, dann heißt f **monoton wachsend (steigend) auf $I_1$**

Gilt für alle $x_1, x_2 \in I_2$, wobei $x_1 < x_2$ ist, auch dass $f(x_1) \geq f(x_2)$ ist, dann heißt f **monoton fallend auf $I_2$**
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

# Defintionsmenge und Funktion
# ----------------------------
a= -5.1 # untere x-Intervallgrenze 
b= 5.1 # obere x-Intervallgrenze
c = -5.1# untere y-Intervallgrenze
d = 5.1 # obere y-Intervallgrenze
x = np.linspace(a, b,1000)
y1= x**3-2*x**2-2*x+3
# ----------------------------

# Einstellung des Graphen
fig=plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, aspect =1)

# Definiton der Haupteinheiten, reele Zahlen ohne die 0 
def major_tick(x, pos):
    if x==0:
        return ""
    return int(x)

# Achsenskalierung
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.xaxis.set_major_formatter(FuncFormatter(major_tick))
ax.yaxis.set_major_formatter(FuncFormatter(major_tick))

# Position der Achsen im Schaubild
ax.spines[['top','right']].set_visible(False)
ax.spines[['bottom','left']].set_position('zero')

# Pfeile für die Achsen
ax.plot((1),(0), ls="", marker= ">", ms=7, color="k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0),(1), ls="", marker= "^", ms=7, color="k", transform=ax.get_xaxis_transform(), clip_on=False)

# Achsenlänge und Beschriftung
ax.set_xlim(a,b)
ax.set_ylim(c, d)
ax.set_xlabel("x", loc="right")
ax.set_ylabel("f(x)", loc="top", rotation=0)

# Kästchen
ax.grid(linestyle="-", which="major",linewidth=0.7, zorder=-10)
ax.grid(linestyle="-", which="minor",linewidth=0.5, zorder=-10)

# Plot der Funktion
ax.plot(x,y1, zorder=10)
#plt.show()
```

Mit Hilfe der Ableitungsfunktion lässt sich die Steigung eines Graphen an einer Stelle bestimmen.

Wenn auf einem Intervall die Werte der Ableitung einer Funktion alle positiv sind, so ist die Funktion

Wenn auf einem Intervall die Werte der Ableitung einer Funktion alle negativ sind, so ist die Funktion

Dies führt zum Monotoniesatz

<div style="border: 1px solid red">

**Monotoniesatz:**
Gegeben ist eine im Intervall I differenzierbare Funktion. Wenn für alle $x\in I$ gilt:

- $f'(x)>0$, dann ist f streng monoton wachsend auf I
- $f'(x)<0$, dann ist f streng monoton fallend auf I

</div>


**Wichtige Bemerkung:**

**Die Umkehrung es Satzes gilt nicht!**


**Gegenbeispiel:** $f(x)=x^3$




```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

# Defintionsmenge und Funktion
# ----------------------------
a= -5.1 # untere x-Intervallgrenze 
b= 5.1 # obere x-Intervallgrenze
c = -5.1# untere y-Intervallgrenze
d = 5.1 # obere y-Intervallgrenze
x = np.linspace(a, b,1000)
y1= x**3
# ----------------------------

# Einstellung des Graphen
fig=plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1, aspect =1)

# Definiton der Haupteinheiten, reele Zahlen ohne die 0 
def major_tick(x, pos):
    if x==0:
        return ""
    return int(x)

# Achsenskalierung
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.xaxis.set_major_formatter(FuncFormatter(major_tick))
ax.yaxis.set_major_formatter(FuncFormatter(major_tick))

# Position der Achsen im Schaubild
ax.spines[['top','right']].set_visible(False)
ax.spines[['bottom','left']].set_position('zero')

# Pfeile für die Achsen
ax.plot((1),(0), ls="", marker= ">", ms=7, color="k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0),(1), ls="", marker= "^", ms=7, color="k", transform=ax.get_xaxis_transform(), clip_on=False)

# Achsenlänge und Beschriftung
ax.set_xlim(a,b)
ax.set_ylim(c, d)
ax.set_xlabel("x", loc="right")
ax.set_ylabel("f(x)", loc="top", rotation=0)

# Kästchen
ax.grid(linestyle="-", which="major",linewidth=0.7, zorder=-10)
ax.grid(linestyle="-", which="minor",linewidth=0.5, zorder=-10)

# Plot der Funktion
ax.plot(x,y1, zorder=10)
#plt.show()
```

Es gilt $f'(x)=2x^3$ und an der Stelle $x=0$ somit $f'(0)=0$

Aber f ist nach der Defintion der Monotonie streng monoton wachsend, denn für alle $x_1, x_2 \in \mathbb{R}$ mit $x_1<x_2$ gilt $(x_1)<f(x_2)$

Wie bei einer Straße lassen sich bei Graphen **Links-** und **Rechtskurven** unterscheiden.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter

# Defintionsmenge und Funktion
# ----------------------------
a= -5.1 # untere x-Intervallgrenze 
b= 5.1 # obere x-Intervallgrenze
c = -5.1# untere y-Intervallgrenze
d = 5.1 # obere y-Intervallgrenze
x = np.linspace(a, b,1000)
y1= np.exp(x)
y2= -np.exp(-x)+3
# ----------------------------

# Einstellung des Graphen
fig=plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,2,1, aspect =1)
ax1 = fig.add_subplot(1,2,2, aspect =1)

# Definiton der Haupteinheiten, reele Zahlen ohne die 0 
def major_tick(x, pos):
    if x==0:
        return ""
    return int(x)

# Achsenskalierung
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_major_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.xaxis.set_major_formatter(FuncFormatter(major_tick))
ax.yaxis.set_major_formatter(FuncFormatter(major_tick))

ax1.xaxis.set_major_locator(MultipleLocator(1))
ax1.xaxis.set_minor_locator(AutoMinorLocator(2))
ax1.yaxis.set_major_locator(MultipleLocator(1))
ax1.yaxis.set_minor_locator(AutoMinorLocator(2))
ax1.xaxis.set_major_formatter(FuncFormatter(major_tick))
ax1.yaxis.set_major_formatter(FuncFormatter(major_tick))

# Position der Achsen im Schaubild
ax.spines[['top','right']].set_visible(False)
ax.spines[['bottom','left']].set_position('zero')

ax1.spines[['top','right']].set_visible(False)
ax1.spines[['bottom','left']].set_position('zero')

# Pfeile für die Achsen
ax.plot((1),(0), ls="", marker= ">", ms=7, color="k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot((0),(1), ls="", marker= "^", ms=7, color="k", transform=ax.get_xaxis_transform(), clip_on=False)

ax1.plot((1),(0), ls="", marker= ">", ms=7, color="k", transform=ax1.get_yaxis_transform(), clip_on=False)
ax1.plot((0),(1), ls="", marker= "^", ms=7, color="k", transform=ax1.get_xaxis_transform(), clip_on=False)

# Achsenlänge und Beschriftung
ax.set_xlim(a,b)
ax.set_ylim(c, d)
ax.set_xlabel("x", loc="right")
ax.set_ylabel("f(x)", loc="top", rotation=0)

ax1.set_xlim(a,b)
ax1.set_ylim(c, d)
ax1.set_xlabel("x", loc="right")
ax1.set_ylabel("f(x)", loc="top", rotation=0)

# Kästchen
ax.grid(linestyle="-", which="major",linewidth=0.7, zorder=-10)
ax.grid(linestyle="-", which="minor",linewidth=0.5, zorder=-10)

ax1.grid(linestyle="-", which="major",linewidth=0.7, zorder=-10)
ax1.grid(linestyle="-", which="minor",linewidth=0.5, zorder=-10)

# Plot der Funktion
plt.subplot(121)
plt.plot(x,y1, zorder=10)

plt.subplot(122)
plt.plot(x,y2, zorder = 10)
#plt.show()
```

Beide Funktionsgraphen, also auch beide Funktionen sind streng monoton wachsend.

**Links:** Die Zunahme wird immer größer. Der Graph ist von f ist also **linksgekrümmt.** Er beschreibt eine Linkskurve.

**Rechts:** Die Zunahme wird immer kleiner. Der Graph von f ist also **rechtsgekrümmt.**  Er beschreibt eine Rechtskurve.

</br>
Wenn eine Funktion f differenzierbar ist, so kann man die Krümmung mit Hilfe der Ableitungsfunktion f' beschreiben.


**Links:** 
- f beschreibt eine Linkskurve
- f' wird immer größer
- f' ist streng monoton wachsend
- f'' ist größer gleich 0

**Rechts:** 
- f beschreibt eine Rechtskurve
- f' wird immer kleiner
- f' ist streng monoton fallend
- f'' ist kleiner gleich 0


Zusammenfassend und auf Intervalle betrachtet führt dies zu folgenden Satz:

<div style="border: 1px solid red">

**Satz:**

Gegeben:

- Funtion f auf einem Intervall I definiert.
- f zweimal auf dem Intervall I differenzierbar.

Es gilt:

- Wenn $f''(x) > 0$ für alle $x \in I$, dann ist der Graph von f auf $I$ **linksgekrümmt.**
- Wenn $f''(x) < 0$ für alle $x \in I$, dann ist der Graph von f auf $I$ **rechtsgekrümmt.**
</div>
